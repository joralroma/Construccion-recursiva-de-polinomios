```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol('x')
```

```python
def buildPolynomials(n):
  polynomials = [x - 2, x**2 + x]

  for i in range(2, n):
    polynomial = (x/2)*polynomials[i-1] - x*polynomials[i-2]*(x/4)
    polynomials.append(polynomial.expand())
  
  return polynomials
```

```python
result = buildPolynomials(50)
```

```python
result[len(result) - 1]
```

### Gráfica


```python
xs = np.linspace(-1, 1, num = 100)
for i in range(1, 5):
  lambda_tn = sp.lambdify(x, result[i], 'numpy')
  ys = lambda_tn(xs)
  
  plt.plot(xs, ys)
```
