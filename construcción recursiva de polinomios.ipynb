```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol('x')
```

```python
def buildPolynomials(n):
  polynomials = [x - 2, x**2 + x]

  for i in range(2, n):
    polynomial = (x/2)*polynomials[i-1] - x*polynomials[i-2]*(x/4)
    polynomials.append(polynomial.expand())
  
  return polynomials
```

```python
result = buildPolynomials(50)
```
